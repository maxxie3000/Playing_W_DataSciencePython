## Exercise 13: Functional Approach - Line Plot

```python
#Generate array of numbers
import numpy as np

x = np.linspace(0, 10, 20)
```

```python
#Create an array and save it to object y
y = x**3
```

```python
#Create the plot
import matplotlib.pyplot as plt

plt.plot(x, y)
plt.show()
```

```python
#Add an s-axis label
plt.xlabel('Linearly Spaced Numbers')
#Add ylable
plt.ylabel('y value')
#Add title 
plt.title('x by x cubed')
#Change the color by using 'k'
plt.plot(x,y,'k')
```

```python
#Change the line characters into a diamond
plt.plot(x,y,'Db')
```

```python
#Connect the diamonds with a solid line 
plt.plot(x, y, 'D-b')
```

## Exercise 14: Functional Approach - Add a second line 

```python
y2 = x ** 2
```

```python
#Connect the diamonds with a solid line (Note: Figure must be before plot)
plt.figure(figsize=(10,5))
plt.plot(x, y, 'D-b', label = 'x cubed')
plt.plot(x, y2, '--r', label = 'x squared')
plt.legend(loc='upper left')
```

## Exercise 15: Creating a bar plot

## Exercise 16: Functional Approach - Histogram

```python
import numpy as np 
#Create sample data
y = np.random.normal(loc=0, scale=0.1, size=100)
```

```python
plt.hist(y, bins=20)

plt.xlabel('y Value')
plt.ylabel('Frequency')
```

```python
#to test whether distribution is normal se Shapiro-Wilk test
from scipy.stats import shapiro 

shap_w, shap_p = shapiro(y)

if shap_p > 0.05:
    normal_yn = 'Fail to reject the null hypothesis. Data is normally distributed'
else: 
    normal_yn = 'Null hypothesis is rejected. Daya is not normally distributed'

plt.hist(y, bins=20)

plt.xlabel('y Value')
plt.ylabel('Frequency')
plt.title(normal_yn)

plt.show()
```

## Exercise 17: Functional Approach - Box-and-Whisker plot

```python
#generate data
import numpy as np

y = np.random.normal(loc=0, scale=0.1, size=100)
```

```python
import matplotlib.pyplot as plt

plt.boxplot(y)
plt.show()
```

```python
#A method to detect number of outliers 
from scipy.stats import shapiro, zscore

shap_w, shap_p = shapiro(y)
#Zscores provides the number of standard deviations the value is from the mean 
y_z_scores = zscore(y)

total_outliers = 0 
for i in range(len(y_z_scores)):
    if abs(y_z_scores[i]) >= 3:
        total_outliers += 1

#Generate title 
if shap_p > 0.05:
    title = f'Normally distributed with {total_outliers} outlier(s)'
else:
    title = f'Not normally distributed with {total_outliers} outlier(s)'
    
plt.boxplot(y)
plt.title(title)
plt.show()
```

## Exercise 18: Scatterplot

```python
#Not creating list, but random integers
import numpy as np

y = np.random.uniform(5, 10, 10)

x = np.random.uniform(100,200,10)
```

```python
import matplotlib.pyplot as plt 

plt.scatter(x, y)

plt.xlabel('Weight')
plt.ylabel('Height')
```

```python
from scipy.stats import pearsonr
correlation_coeff, p_value = pearsonr(x, y)
print(correlation_coeff)

# Set up some logic
if correlation_coeff == 1.00:
    title = 'There is a perfect positive linear relationship (r = {0:0.2f}).'.format(correlation_coeff)
elif correlation_coeff >= 0.8:
    title = 'There is a very strong, positive linear relationship (r = {0:0.2f}).'.format(correlation_coeff)
elif correlation_coeff >= 0.6:
    title = 'There is a strong, positive linear relationship (r = {0:0.2f}).'.format(correlation_coeff)
elif correlation_coeff >= 0.4:
    title = 'There is a moderate, positive linear relationship (r = {0:0.2f}).'.format(correlation_coeff)
elif correlation_coeff >= 0.2:
    title = 'There is a weak, positive linear relationship (r = {0:0.2f}).'.format(correlation_coeff)
elif correlation_coeff > 0:
    title = 'There is a very weak, positive linear relationship (r = {0:0.2f}).'.format(correlation_coeff)
elif correlation_coeff == 0:
    title = 'There is no linear relationship (r = {0:0.2f}).'.format(correlation_coeff)
elif correlation_coeff <= -0.8:
    title = 'There is a very strong, negative linear relationship (r = {0:0.2f}).'.format(correlation_coeff)
elif correlation_coeff <= -0.6:
    title = 'There is a strong, negative linear relationship (r = {0:0.2f}).'.format(correlation_coeff)
elif correlation_coeff <= -0.4:
    title = 'There is a moderate, negative linear relationship (r = {0:0.2f}).'.format(correlation_coeff)
elif correlation_coeff <= -0.2:
    title = 'There is a weak, negative linear relationship (r = {0:0.2f}).'.format(correlation_coeff)
else: 
    title = 'There is a very weak, negative linear relationship (r = {0:0.2f}).'.format(correlation_coeff)
```

```python
import matplotlib.pyplot as plt 

plt.scatter(x, y)

plt.xlabel('Weight')
plt.ylabel('Height')
plt.title(title)
plt.show()
```

## Exercise 19: Single Line Plot using Subplots

```python
import numpy as np

x = np.linspace(0, 10, 20)

y = x ** 3
```

```python
import matplotlib.pyplot as plt 
#Using this approach makes Fig object callable in other cells
fig, axes = plt.subplots()
```

```python
axes.plot(x,y)

axes.plot(x, y, 'D--k')

axes.set_xlabel('Linearly Spaced Numbers')
axes.set_ylabel('y Value')
axes.set_title('As x increases, y increases by x cubed')

fig
```

## Exercise 20: Multiple Line Plots Using Subplots 

```python
import numpy as np

x = np.linspace(0, 10, 20)

y = x**3

y2 = x**2
```

```python
import matplotlib.pyplot as plt 

fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
#Left figure
axes[0].plot(x,y)

axes[0].set_xlabel('Linearly Spaced Numbers')
axes[0].set_ylabel('y Value')
axes[0].set_title('x by x cubed')

#Right figure 
axes[1].plot(x, y2)

axes[1].set_xlabel('Linearly Spaced Numbers')
axes[1].set_ylabel('y Value')
axes[1].set_title('x by x Squared')

fig.tight_layout()

fig
```

## Activity 4: Multiple Plot Types Using Subplots 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np

df = pd.read_csv('Data/Items_Sold_by_Week.csv')
df1 = pd.read_csv('Data/Weight_by_Height.csv')
print(df.columns)

r_array = np.random.normal(loc=0, scale=0.1, size = 100)

#Generate six empty figure
fig, axes = plt.subplots(nrows = 3, ncols = 2, figsize=(20,30))

axes[0,0].plot(df['Week'], df['Items_Sold'])
axes[0,1].bar(df['Week'], df['Items_Sold'])
axes[1,0].barh(df['Week'], df['Items_Sold'])

axes[0,0].set_xlabel('Week')
axes[0,1].set_xlabel('Week')
axes[1,0].set_xlabel('Items sold')

axes[0,0].set_ylabel('Items sold')
axes[0,1].set_ylabel('Items sold')
axes[1,0].set_ylabel('Week')

axes[0,0].set_title('Line')
axes[0,1].set_title('Bar')
axes[1,0].set_title('BarH')

axes[1,1].hist(r_array)
axes[2,1].boxplot(r_array)

axes[1,1].set_xlabel('Values')
axes[1,1].set_ylabel('Frequency')
axes[1,1].set_title('Histogram')

axes[2,1].set_ylabel('Values')
axes[2,1].set_title('Box Plot')

axes[2,0].scatter(df1['Weight'], df1['Height'])

axes[2,0].set_xlabel('Weight')
axes[2,0].set_ylabel('Height')
axes[2,0].set_title('Weight by Height')

fig.tight_layout()

fig.show()
```
