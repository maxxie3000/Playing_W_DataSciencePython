# Activity 1: Pre-Processing

```python
import pandas as pd 

df = pd.read_csv('Data/Banking_Marketing.csv', header = 0)
```

```python
#Check if there are null values and 
null_ = df.isna().any()
dtypes = df.dtypes
info = pd.concat([null_, dtypes], axis = 1, keys = ['Null', 'Dtypes'])
print(info)
```

```python
#Fill missing values 
df.contact.fillna(df.contact.mode()[0], inplace = True)

df.duration.fillna(df.duration.median(), inplace = True)

df.dropna(subset=['age'], inplace = True)

df
```

```python
df.education.value_counts()
```

```python
df['education'] =  df.education.replace({'basic.9y': 'basic', 'basic.6y': 'basic',
                                         'basic.4y': 'basic'})
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

OneHotEncoder = OneHotEncoder(sparse = False)

cat_columns = df.select_dtypes(exclude=[np.number]).columns
num_columns = df.select_dtypes([np.number]).columns

onehot_encoded = OneHotEncoder.fit_transform(df[cat_columns])

df_enc = pd.DataFrame(onehot_encoded, columns = OneHotEncoder.get_feature_names(cat_columns))

df_enc = pd.concat([df_enc, df[num_columns]], axis = 1)
```

```python
from sklearn.model_selection import train_test_split

#Splitting data
X = df_enc.drop('y', axis = 1)
y = df_enc['y']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```
