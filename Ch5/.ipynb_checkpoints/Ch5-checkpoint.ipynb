## Exercise 44: Using the XGBoost library to Perform Classification

```python
import pandas as pd 
import xgboost as xgb
from sklearn.metrics import accuracy_score 

data = pd.read_csv('data/wholesale-data.csv')
```

```python
data.head()
```

```python
X = data.copy()
X.drop('Channel', inplace = True, axis = 1)
y = data.Channel
```

```python
X_train, X_test = X[:int(X.shape[0]*0.8)].values, X[int(X.shape[0]*0.8):].values

y_train, y_test = y[:int(y.shape[0]*0.8)].values, y[int(y.shape[0]*0.8):].values
```

```python
#Convert to DMatrix, internal data structure of XGBoost 
train = xgb.DMatrix(X_train, label = y_train)
test = xgb.DMatrix(X_test, label = y_test)
```

```python
param = {'max_depth':6, 'eta':0.1, 'verbosity':0, 'objective':'multi:softmax', 'num_class': 3}
num_round = 5
model = xgb.train(param, train, num_round)
```

```python
preds = model.predict(test)
acc = accuracy_score(y_test, preds)
print(f'Accuracy : {acc*100:.5f} %')
```

## Activity 14: Training and Predicting the income of a person

```python
import pandas as pd 

df = pd.read_csv('data/adult-data.csv', names=['age', 'workclass', 'education-num', 'occupation', 'capital-gain', 
                                               'capital-loss', 'hoursper-week', 'income'])
```

```python
df.head()
```

```python
df.occupation.value_counts()
```

```python
df.income.value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

#Using full dataset!
#Get category columns
data_column_category = df.select_dtypes(exclude=[np.number]).columns
#Performing OneHot encoding
onehot_encoder = OneHotEncoder(sparse=False)
onehot_encoded = onehot_encoder.fit_transform(df[data_column_category])

#Creating new dataframe with encoded data
onehot_encoded_frame = pd.DataFrame(onehot_encoded, columns = onehot_encoder.get_feature_names(data_column_category))
#Getting number columns
data_column_number = df.select_dtypes(include=[np.number]).columns
#Creating new dataframe
data_onehot_encoded_data = pd.concat([onehot_encoded_frame, df[data_column_number]], axis = 1)
```

```python
data_onehot_encoded_data['income'] = data_onehot_encoded_data['income_ <=50K'].apply(lambda x: 0 if x == 1 else 1)
```

```python
data = data_onehot_encoded_data.drop(['income_ <=50K', 'income_ >50K'], axis = 1)
```

```python
X = data.copy()
X.drop('income', inplace = True, axis = 1)
y = data.income
```

```python
X_train, X_test = X[:int(X.shape[0]*0.8)].values, X[int(X.shape[0]*0.8):].values

y_train, y_test = y[:int(y.shape[0]*0.8)].values, y[int(y.shape[0]*0.8):].values
```

```python
#Convert to DMatrix, internal data structure of XGBoost 
train = xgb.DMatrix(X_train, label = y_train)
test = xgb.DMatrix(X_test, label = y_test)
```

```python
param = {'max_depth':12, 'eta':0.5, 'verbosity':2, 'objective':'binary:hinge', 'eval_metric':'aucpr'}
num_round = 5
model = xgb.train(param, train, num_round)
```

```python
preds = model.predict(test)
acc = accuracy_score(y_test, preds)
print(f'Accuracy : {acc*100:.5f} %')
```

## Exercise 46: Crating a Python Pcript that predicts Based on Real-time Input

```python
import pandas as pd 
import numpy as np

data = pd.read_csv('data/adult-data.csv', names=['age', 'workclass', 'education-num', 'occupation', 'capital-gain', 
                                               'capital-loss', 'hoursper-week', 'income'])
```

```python
data.occupation.value_counts()
```

```python
#Strip away all trailing spaces 
data[['workclass', 'occupation', 'income']] = data[['workclass', 'occupation', 'income']].apply(lambda x:
                                                                                               x.str.strip())

#Convert categorical variable from strings to integers
from sklearn.preprocessing import LabelEncoder 
from collections import defaultdict 

label_dict = defaultdict(LabelEncoder)
data[['workclass', 'occupation', 'income']] = data[['workclass', 'occupation', 'income']].apply(lambda x:
                                                                                                label_dict[x.name]\
                                                                                                .fit_transform(x))

#Save label encoder in a pickle file for future use 
import pickle 
with open('income_labels.pkl', 'wb') as f:
    pickle.dump(label_dict, f, pickle.HIGHEST_PROTOCOL)
X = data.copy()
X.drop("income", inplace = True, axis = 1)
Y = data.income
```

```python
X = data.copy()
X.drop("income", inplace = True, axis = 1)
Y = data.income
```

```python
X_train, X_test = X[:int(X.shape[0]*0.8)].values, X[int(X.shape[0]*0.8):].values
Y_train, Y_test = Y[:int(Y.shape[0]*0.8)].values, Y[int(Y.shape[0]*0.8):].values
```

```python
train = xgb.DMatrix(X_train, label=Y_train)
test = xgb.DMatrix(X_test, label=Y_test)
```

```python
param = {'max_depth':7, 'eta':0.1, 'verbosity':1, 'objective':'binary:hinge'}
num_round = 50
model = xgb.train(param, train, num_round)
```

```python
preds = model.predict(test)
```

```python
model.save_model('income-model.model')
```

```python
import xgboost as xgb
loaded_model = xgb.Booster()
loaded_model.load_model('income-model.model')
def load_obj(file):
    with open(file+'.pkl', 'rb') as f:
        return pickle.load(f)

label_dict = load_obj('income_labels')
```

```python
#Read the input 
age = input('Enter Age: ')
workclass = input('Enter workclass: ')
education_num = input('Enter education_num: ')
occupation = input('Enter occupation: ')
capital_gain = input('Enter capital_gain: ')
capital_loss = input('Enter capital_loss: ')
hours_per_week = input('Enter hours_per_week: ')
```

```python
data_list = [age, workclass, education_num, occupation, capital_gain, capital_loss, hours_per_week]
data = pd.DataFrame([data_list])
data.columns = ['age', 'workclass', 'education_num', 'occupation', 'capital_gain', 'capital_loss', 'hours_per_week']

#Preprocess the data 
data[['workclass', 'occupation']] = data[['workclass', 'occupation']].apply(lambda x:
                                                                                    label_dict[x.name]\
                                                                                    .transform(x))

data = data.astype(int)
data_xgb = xgb.DMatrix(data)
pred = loaded_model.predict(data_xgb)

income = label_dict['income'].inverse_transform([int(pred[0])])
print(income)
```

## Exercise 47: Installing the Keras library for Python and Using it to perform Classification

```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from sklearn.metrics import accuracy_score
from keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/wholesale-data.csv')
data.info()
```

```python
X = data.copy()
X.drop("Channel", inplace = True, axis = 1)
Y = data.Channel
```

```python
X_train, X_test = X[:int(X.shape[0]*0.8)].values, X[int(X.shape[0]*0.8):].values
Y_train, Y_test = Y[:int(Y.shape[0]*0.8)].values, Y[int(Y.shape[0]*0.8):].values
```

```python
#Create the neural network
model = Sequential()
model.add(Dense(units=8, activation='relu', input_dim=7))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
#Compile the model
model.compile(loss='binary_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
model.summary()
```

```python
#Fitting the model
model.fit(X_train, Y_train, epochs=5, batch_size=8)
```

```python
#Get predictions
preds = model.predict(X_test, batch_size=128)
#Obtain accuracy
acc = accuracy_score(Y_test, preds.astype(int))
print(f'Accuracy {acc*100:.2f}')
```
